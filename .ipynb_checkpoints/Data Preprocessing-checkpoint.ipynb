# Toronto Bike Share Project


The files obtained were stored as seperate CSV based on each month. The years range from 2021 to the now current, 2023. The files needs to be first combined into one CSV for easier use across Python, Tableau and MySql. 

```python
# importing necessary libraries
import numpy as np
import pandas as pd
import os
```

```python
# combining all the csv files into one file

directory = 'data/'
# an empty DataFrame to hold all the data
combined_data = pd.DataFrame()

# iterate through each csv file in the directory
for filename in os.listdir(directory):
    if filename.endswith('csv'):
        filepath = os.path.join(directory, filename)
        print(filepath)
        # read the csv file into a DataFrame
        df = pd.read_csv(filepath, encoding='latin-1', skipinitialspace=True)
        df.columns = df.columns.str.replace('ï»¿', '')
        print(df.columns)
        # append the DataFrame to the combined_data DataFrame
        combined_data = combined_data.append(df, ignore_index=True)
        
# Save the combined data to a new Excel file
combined_data.to_csv('CombinedData.csv', index=False)
```

```python
# setting df to combined_data for easier access
df = combined_data
```

The time format needs to be changed to standard datetime (YYYY-MM-DD) for import into MySql.

```python
from datetime import datetime
# Converting times to datetime format
df['Start Time'] = df['Start Time'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %H:%M').strftime('%Y-%m-%d %H:%M'))
df['End Time'] = df['End Time'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y %H:%M').strftime('%Y-%m-%d %H:%M'))
```
